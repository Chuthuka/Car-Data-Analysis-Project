**2. ----Value Counts of Make, Model, Type, Origin----**

```python
data.head(2)
```

**2.1 "Make" Column**

```python
data['Make'].value_counts()
```

```python
plt.figure(figsize=(12, 8))
data['Make'].value_counts().plot(kind='bar', color='#B19CD9')
plt.title('Count By Vehicle Make', pad=15, fontweight='bold', fontsize=15)     
plt.ylabel('Number of Vehicles', fontweight='bold',fontsize=15, labelpad=15) 
plt.xlabel('Vehicle Type', fontweight='bold', fontsize=15,  labelpad=15)

plt.show();
```

**2.3 "Type" Column**

```python
data['Type'].value_counts().plot(kind = 'bar')
plt.title('Count By Vehicle Types', pad=15, fontweight='bold')     
plt.ylabel('Number of Vehicles', fontweight='bold',fontsize=12, labelpad=15) 
plt.xlabel('Vehicle Type', fontweight='bold', fontsize=12,  labelpad=15)

plt.show();
```

**2.4 "Origin" Column**

```python
data['Origin'].value_counts()
```

```python
counts = data['Origin'].value_counts()
labels = counts.index
sizes = counts.values


colors = ['#FFA500', '#FF4500', '#FF0000']  

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors[:len(sizes)],
        autopct='%1.1f%%', startangle=90,
        textprops={'fontsize': 14, 'fontweight': 'bold'}, explode=[0.05] + [0]*(len(sizes)-1))

plt.title('Vehicle Origins Distribution', fontsize=18, fontweight='bold')
plt.axis('equal')  # Keep it circular
plt.show()
```
